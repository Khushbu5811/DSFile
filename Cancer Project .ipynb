# Importing the libraries and dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("data.csv")
data.head()
```

# To get all the column with null value column

```python
data.isna().sum()
```

```python
# returns the size of dataset
data.shape
```

```python
# remove the column
df= data.dropna(axis = 1)
```

```python
# shape of data set after removing null colum
df.shape
```

# Statistical summary

```python
df.describe()
```

# Dealing with categorical data

```python
# get the count of malignant<M> and benign <B> cells
df['diagnosis'].value_counts()
```

# Visualization

```python
sns.catplot(x='diagnosis', kind='count', data=df)
plt.title('Count Plot of Diagnosis')
plt.show()
```

```python
# Create a pie chart to visualize class distribution
class_counts = data["diagnosis"].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Class Distribution of 'diagnosis'")
plt.show()
```

# One hot encoding

```python
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})
```

```python
df.head()
```

```python
sns.pairplot(df.iloc[:,1:5],hue = "diagnosis")           # iloc to give range
```

# Correlation matrix 

```python
cor = df.corr()
cor
```

# Heatmap

```python
# Visualize the correlation
plt.figure(figsize=(25,23))
sns.heatmap(cor, annot= True, linewidths= 0.3 ,linecolor = "black", fmt = ".2f")
plt.title('Correlation Heatmap')
plt.show()
```

# Removing Insignificant columns

```python
dataset=df.drop(["id",'fractal_dimension_mean','texture_se','smoothness_se','symmetry_se'],axis=1)
```

```python
dataset.head()
```

```python
dataset.shape
```

# Splitting the datatset train and test set

```python
X = dataset.iloc[:,1:27].values
Y = dataset.iloc[:,0].values
```

```python
print(X)
```

```python
print(Y)
```

```python
# sklearn use to break the data
# Spliting the data into training and test dataset
# random state is used for not to get bise

from sklearn.model_selection import train_test_split
X_train , X_test , Y_train , Y_test = train_test_split(X, Y, test_size = 0.2 , random_state = 0)
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
X_test.shape
```

```python
 Y_test .shape
```

# Features scaling


```python
from sklearn.preprocessing import StandardScaler
X_train = StandardScaler().fit_transform(X_train)
X_test = StandardScaler().fit_transform(X_test)
```

# Building the model

# 1.  Logistic Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,f1_score,precision_score,recall_score
```

```python
logistic_model = LogisticRegression(random_state=0)
logistic_model.fit(X_train, Y_train)
```

```python
Y_pred = logistic_model.predict(X_test)
print("Predicted values :")
print(Y_pred)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,f1_score,precision_score,recall_score
```

```python
accuracy_score = accuracy_score(Y_test,Y_pred)
f1_score = f1_score(Y_test,Y_pred)
precision_score = precision_score(Y_test,Y_pred)
recall_score = recall_score(Y_test,Y_pred)
```

# DataFrame of result

```python
Results = pd.DataFrame([['Logistic Regression',accuracy_score,f1_score,precision_score,recall_score]],
                    columns=['Model','Accuracy','f1 score','precision','recall'])
```

```python
Results
```

```python
cm = confusion_matrix(Y_test,Y_pred)
print(cm)
```

```python
sns.heatmap(cm,annot=True,
            xticklabels=['Malignant','Benign'],
            yticklabels=['Malignant','Benign'])
plt.ylabel('Prediction',fontsize=13)
plt.xlabel('Actual',fontsize=13)
plt.title('Confusion Matrix',fontsize=17)
plt.show()
```

```python
print("Predicted values :")
print(Y_pred)
print("Actual values :")
print(Y_test)
```

# Cross Validation

```python
from sklearn.model_selection import cross_val_score
accuracies= cross_val_score(estimator= logistic_model, X=X_train, y=Y_train, cv=10)
print("accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation : {:.2f}%".format(accuracies.std()*100))
lrcv=accuracies.mean()*100
print(lrcv)
```
